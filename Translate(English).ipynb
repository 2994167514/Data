```python
from qiskit import QuantumCircuit, Aer, execute
qasm_sim = Aer.get_backend('qasm_simulator')
```

```python
print("1. Sara sees Bob")
qc = QuantumCircuit(5,4)
qc.rx(0.0,0)
qc.h(1)
qc.h(2)
qc.h(3)
qc.rx(0.2,4)

qc.rz(0.002,0)
qc.crz(0.012,1,2)
qc.rz(0.202,4)

qc.rx(0.004, 0)
qc.h(1)
qc.crz(0.014,2,3)
qc.rx(0.204,4)

qc.h(2)
qc.h(3)

qc.crz(0.016,1,2)

qc.cx(0,1)
qc.crz(0.018,2,3)
qc.barrier()

qc.measure(1,1)
qc.barrier()

qc.h(0)
qc.cx(3,4)
qc.barrier()

qc.measure(0,0)
qc.measure(4,3)
qc.barrier()

qc.h(3)

qc.measure(3,2)

result = execute(qc, qasm_sim).result()
counts = result .get_counts()
qc.draw(output='mpl', scale=0.9)
```
